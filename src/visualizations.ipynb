```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../data/derived/athletes_height_weight.csv")
```

```python
plt.hist(df[df['Sex'] == 'M']['Age'], bins=20, color='blue', edgecolor='black', alpha=0.7, label='Male')
plt.hist(df[df['Sex'] == 'F']['Age'], bins=20, color='pink', edgecolor='black', alpha=0.7, label='Female')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Athletes', fontweight = 'bold')
plt.legend()
plt.grid(False)
plt.show()
```

```python

plt.hist(df[df['Sex'] == 'M']['Height'], bins=20, color='blue', edgecolor='black', alpha=0.7, label='Male')
plt.hist(df[df['Sex'] == 'F']['Height'], bins=20, color='pink', edgecolor='black', alpha=0.7, label='Female')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.title('Height Distribution of Athletes', fontweight = 'bold')
plt.legend()
plt.grid(False)
plt.show()
```

```python
plt.hist(df[df['Sex'] == 'M']['Weight'], bins=20, color='blue', edgecolor='black', alpha=0.7, label='Male')
plt.hist(df[df['Sex'] == 'F']['Weight'], bins=20, color='pink', edgecolor='black', alpha=0.7, label='Female')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.title('Weight Distribution of Athletes', fontweight = 'bold')
plt.legend()
plt.grid(False)
plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# AGE DISTRIBUTION PLOT 
axs[0].hist(df[df['Sex'] == 'M']['Age'], bins=20, color='#ADD8E6', edgecolor='black', alpha = 0.8, label='Male')
axs[0].hist(df[df['Sex'] == 'F']['Age'], bins=20, color='#FFB6C1', edgecolor='black', alpha = 0.8, label='Female')
axs[0].set_xlabel('Age')
axs[0].set_ylabel('Frequency')
axs[0].set_title('Age Distribution', fontweight='bold')
axs[0].legend()

# HEIGHT DISTRIBUTION PLOT 
axs[1].hist(df[df['Sex'] == 'M']['Height'], bins=20, color='#6495ED', edgecolor='black', alpha = 0.8, label='Male')
axs[1].hist(df[df['Sex'] == 'F']['Height'], bins=20, color='#FF69B4', edgecolor='black', alpha = 0.8, label='Female')
axs[1].set_xlabel('Height (cm)')
axs[1].set_ylabel('Frequency')
axs[1].set_title('Height Distribution', fontweight='bold')
axs[1].legend()

# WEIGHT DISTRIBUTION PLOT 
axs[2].hist(df[df['Sex'] == 'M']['Weight'], bins=20, color='#4169E1', edgecolor='black', alpha = 0.8, label='Male')
axs[2].hist(df[df['Sex'] == 'F']['Weight'], bins=20, color='#FF1493', edgecolor='black', alpha = 0.8, label='Female')
axs[2].set_xlabel('Weight (kg)')
axs[2].set_ylabel('Frequency')
axs[2].set_title('Weight Distribution', fontweight='bold')
axs[2].legend()

plt.tight_layout()
plt.show()
```
